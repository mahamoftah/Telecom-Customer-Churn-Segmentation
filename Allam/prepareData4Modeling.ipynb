```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Internship/Neuronetix/Telecom Customers Churn.csv')
data.info()
```

```python
data.isna().sum()
```

```python
dic = {}
for column in data.columns:
  dic[column] = data[column].nunique()

pd.DataFrame(dic, index=['unique values']).transpose()
```

```python
data.head()
```

```python
data.drop(columns=['customerID'], inplace=True)
data.replace({'Yes': 1, 'No': 0}, inplace=True)
data.replace({'Male': 1, 'Female': 0}, inplace=True)
data.replace({'No phone service': -1, 'No internet service': -1}, inplace=True)
data.replace({'DSL': 1, 'Fiber optic': 2, 'No': 0}, inplace=True)
data.replace({'Month-to-month': 1, 'One year': 12, 'Two year': 24}, inplace=True)
data.replace({'Electronic check': 1, 'Mailed check': 2, 'Bank transfer (automatic)': 3, 'Credit card (automatic)': 4}, inplace=True)

data.head()
```

```python
data.info()
```

```python
dic = {}
for column in data.columns:
  dic[column] = data[column].nunique()

pd.DataFrame(dic, index=['unique values']).transpose()
```

```python
data.to_csv('/content/drive/MyDrive/Colab Notebooks/Internship/Neuronetix/telecom_churn_processed4modeling.csv', index=False)
```
